```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mlt
import math
import json
mlt.rcParams['figure.figsize'] = [24, 12]
```

```python
data = open("./tracker.jsonlines", "r").readlines()
```

```python
df = pd.DataFrame([json.loads(d) for d in data])
```

```python
df = df[["openSpots", "province", "timestamp"]]
```

```python
df[df["province"] == "MI"].groupby(["timestamp"]).sum("openSpots").reset_index(level=0).plot.line(x="timestamp", y="openSpots")
```
